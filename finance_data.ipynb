**Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks.
Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
frames = [df['BAC'],df['C'],df['GS'], df['JPM'],  df['MS'] , df['WFC'] ]
bank_stocks = pd.concat(frames , keys= tickers, axis=1)
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info' )
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick + "Return"] = bank_stocks[tick]['Close'].pct_change()
```

Moving Averages
Let's analyze the moving averages for these stocks in the year 2008.

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
bank_stocks['2008']['BAC']['Close'].rolling(window=30).mean().plot(figsize=(12,4), label="30day average")
bank_stocks['2008']['BAC']['Close'].plot(label="closing price").legend()
```
